```python
import serial
import time
```

```python
portName = "COM3"
baudrate = 57600
bytesize = 7
startbits = 1
stopbits = 1
parity = "O"
timeout = 2

def readLS224():
    lakeShore224 = serial.Serial(port=portName, baudrate=baudrate, bytesize=bytesize,
                                 stopbits=stopbits, parity=parity, timeout=timeout)
    twriteString = b"KRDG? " + bytes("A", encoding='UTF-8') + b"\n"
    lakeShore224.write(twriteString)
    temp = float(lakeShore224.read(10))
    vwriteString = b"SRDG? " + bytes("A", encoding='UTF-8') + b"\n"
    lakeShore224.write(vwriteString)
    volts = float(lakeShore224.read(10))
    lakeShore224.close()
    return temp, volts
```
